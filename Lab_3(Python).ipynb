```python
class Fraction(object):
    def __init__(self, num, den):
        self.__num = num
        self.__den = den
        self.reduce()
        
    def __str__(self):
        return "%d/%d" % (self.__num, self.__den)
    
    def reduce(self):
        g = Fraction.gcd(self.__num, self.__den)
        self.__num /= g
        self.__den /= g
    
    @staticmethod
    def gcd(n, m):
        if m == 0:
            return n
        else:
            return Fraction.gcd(m, n % m)
        
    # Operators #
    def __neg__(self):
        return f"{int(-self.__num)}/{int(self.__den)}"
    
    def __invert__(self):
        return f"{int(self.__den)}/{int(self.__num)}"
    
    def __pow__(self, val):
        return f"{int(self.__num**val)}/{int(self.__den**val)}"
    
    def __float__(self):
        return self.__num/self.__den
    
    def __int__(self):
        return int(self.__num/self.__den)
```

```python
frac = Fraction(7, 2)
print(-frac)       # выводит -7/2
print(~frac)       # выводит 2/7
print(frac**2)     # выводит 49/4
print(float(frac)) # выводит 3.5
print(int(frac))   # выводит 3
```

```python
'''
Задание №2

Напишите классы «Книга» (с обязательными полями: название, автор,
код), «Библиотека» (с обязательными полями: адрес, номер) и
корректно свяжите их. Код книги должен назначаться автоматически
при добавлении книги в библиотеку (используйте для этого
статический член класса). Если в конструкторе книги указывается в
параметре пустое название, необходимо сгенерировать исключение
(например, ValueError). Книга должна реализовывать интерфейс
Taggable с методом tag(), который создает на основе строки набор тегов
(разбивает строку на слова и возвращает только те, которые
начинаются с большой буквы). Например, tag() для книги с названием
‘War and Peace’ вернет список тегов [‘War’, ‘Peace’]. Реализуйте классы
таким образом, чтобы корректно выполнялся следующий код:

lib = Library(1, ’51 Some str., NY’)
lib += Book(‘Leo Tolstoi’, ‘War and Peace’)
lib += Book(‘Charles Dickens’, ‘David Copperfield’)

for book in lib:
 
    # вывод в виде: [1] L.Tolstoi ‘War and Peace’
    print(book)

    # вывод в виде: [‘War’, ‘Peace’]
    print(book.tag())
'''
None
```

```python
import re

# Класс Book #
class Book(object):  
    
    __slots__ = ['__name_book', '__author', '__unique_code']
    
    # Метод для инициализации переменных класса Book #
    def __init__(self, author, name_book):
        
        # Проверяем на наличие наименования книги #
        if (name_book == '') or (type(name_book) != str):  # Если не указано название книги #
            # Выбрасываем исключение #
            raise ValueError
        else:
            # Сохраняем имя #
            self.__name_book = name_book
            
        # Тернарный оператор для назначения имени исходя из условия #
        self.__author = 'Unknown' if (author == '') or (type(author) != str) else author
        # Устанавливаем для новой книги значение уникального кода в None #
        self.__unique_code = None

    # Реализация интерфейса Taggable с методом tag() #
    def tag(self):  
        return re.compile('[A-Z][a-z]+').findall(self.__name_book)  # Ищем слова с большой буквы и возвращаем их список

    # Установка кода книги #
    def set_code(self, unique_code): 
        self.__unique_code = unique_code

    # Приводим к типу string (для вывода) #
    def __str__(self):
        full_name = self.__author.split()[0][0]+'.'+' '.join(self.__author.split()[1:])
        return "[%s] %s '%s'" % (str(self.__unique_code), full_name, self.__name_book)


# Класс Library
class Library(object):
    # Статическая переменная для хранения кода книги #
    unique_code = 1

    __slots__ = ['__number', '__address', '__books']
    
    # Метод для инициализации переменных класса #
    def __init__(self, number, address):
        self.__number = number
        self.__address = address
        self.__books = []

    # Метод для переопределения оператора += #
    def __iadd__(self, book):  
        # Устанавливаем код в книгу
        book.set_code(Library.unique_code)
        Library.unique_code += 1
        # Добавляем книгу в список #
        self.__books.append(book)
        # Возвращаем обработанную библиотеку #
        return self
    
    # Метод для переопределения оператора + #
    def __add__(self, book):
        # Устанавливаем код в книгу
        book.set_code(Library.unique_code) 
        Library.unique_code += 1
        # Добавляем книгу в список #
        self.__books.append(book)
        # Возвращаем обработанную библиотеку #
        return self

    # Преобразовываем в итерируемый объект #
    def __iter__(self):
        return iter(self.__books)
```

```python
lib = Library(1, '51 Some str., NY')
lib += Book('Leo Tolstoi', 'War and Peace')
lib = lib + Book('Uilyam Shekspir', 'A Midsummer Night\'s Dream')
lib += Book('Charles Dickens', 'David Copperfield')

for book in lib:
 
    # вывод в виде: [1] L.Tolstoi ‘War and Peace’
    print(book)

    # вывод в виде: [‘War’, ‘Peace’]
    print(book.tag())
```

```python
class LogFrame(wx.Frame):
    ''' Класс окна отображения лога '''
    
    __log_file = "script18.log"
    
    __slots__ = ['__lines']

    def __init__(self):
        wx.Frame.__init__(
            self, None, -1, "Искатель строк", size=(500, 400))

        if not os.path.isfile(self.__log_file):
            wx.MessageBox('Файл лога не найден. Файл будет создан автоматически', 'Info', wx.OK | wx.ICON_INFORMATION)
            with open('script18.log', 'a+'):
                pass

        self.create_menu()
        self.statusbar = self.CreateStatusBar()
        self.statusbar.SetFieldsCount(2)
        self.statusbar.SetStatusWidths([-3, -2])

        self.list_box = wx.ListBox(self, -1, (20, 20), (80, 120),
                              [], wx.LB_SINGLE)
        self.list_box.Clear()
        
    def menu_data(self):
        data = (("&Файл",
                 ({"&Открыть...": self.on_open_file},)),
                ("&Лог",
                 ({"&Сохранить новый...": self.on_save_new_log},
                  {"&Добавить в лог": self.append_to_log},
                  {"&Просмотр лога": self.on_show_log})
                 ))
        return data

    def create_menu(self):
        menu = wx.MenuBar()

        for item in self.menu_data():
            menu_item = self.create_sub_menu(item[1])
            menu.Append(menu_item, item[0])

        self.SetMenuBar(menu)

    def create_sub_menu(self, itemgroup):
        groupmenu = wx.Menu()

        for item in itemgroup:
            title, handler = list(item.items())[0]
            menu_item = groupmenu.Append(-1, title)
            if handler:
                self.Bind(wx.EVT_MENU, handler, menu_item)

        return groupmenu

    def on_open_file(self, event):
        with wx.FileDialog(self, message="Выберите файл", defaultFile="", wildcard="*.*",
                       style=wx.FD_OPEN | wx.FD_FILE_MUST_EXIST) as fileDialog:
            if fileDialog.ShowModal() == wx.ID_CANCEL:
                return     # the user changed their mind

            # Proceed loading the file chosen by the user
            pathname = fileDialog.GetPath()
            size = os.path.getsize(pathname)
            
            self.statusbar.SetStatusText(f"Обработал файл:{pathname}")
            self.statusbar.SetStatusText(f"{'{0:,}'.format(size).replace(',', ' ')} байт", 1)

            current_time = datetime.datetime.now().strftime('%d.%m.%Y %H:%M:%S')
            
            self.list_box.Append(f"Файл {pathname} был обработан {current_time}:")
            self.list_box.Append('')
            
            result_list = self.regex_substrings(pathname)
            self.list_box.Append(result_list)
            
    def regex_substrings(self, filename):
        result_list = []
        with open(filename, 'r', encoding='utf-8') as file_handler:
            for line_number, line in enumerate(file_handler):
                required = findall(r'\(\d{3}\)\d{7}', line)
                required += findall(r'\(\d{3}\)\d{3}-\d{2}-\d{2}', line)
                count_out = int((ln(len(line))/ln(10))+1)

                for i in range(len(required)):
                    result_list.append(f"Строка {line_number + 1}, позиция {line.find(required[i]):{count_out}} : найдено '{required[i]}'")
            result_list.append('')
        return result_list
    
    def on_save_new_log(self, event):
        with wx.FileDialog(self, message="Выберите файл", defaultFile="", wildcard="*.*",
                       style=wx.FD_OPEN | wx.FD_FILE_MUST_EXIST) as fileDialog:
            if fileDialog.ShowModal() == wx.ID_CANCEL:
                return     # the user changed their mind
            pathname = fileDialog.GetPath()
            self.append_to_log(self, pathname)
    
    def append_to_log(self, event, path = ''):
        log_list = []
        
        if path == '':
            path = self.__log_file
        
        print(path)
        
        [log_list.append(self.list_box.GetString(i)) for i in range(self.list_box.GetCount())] 
        
        skip = True
        with open(path, "a") as file:
            for  line in log_list:
                if line != '' and skip:
                    skip = False
                if not skip:
                    file.write(line + '\n')
    
    def on_show_log(self, event):
        dlg = wx.MessageDialog(self, 'Вы действительно хотите открыть лог?\nДанные последних поисков будут потеряны!', 'Вопрос',
             wx.YES_NO | wx.NO_DEFAULT | wx.ICON_QUESTION)
        res = dlg.ShowModal()
        
        if res == wx.ID_YES:
            self.on_open_log_file()
        elif res == wx.ID_NO:
            return
        
    def on_open_log_file(self): 
        with open(self.__log_file) as file:
            self.__lines = [line.rstrip() for line in file]
        
        self.list_box.Clear()

        self.list_box.Append(self.__lines)
```

```python
'''
Задание №4:

Напишите простой класс StringFormatter для форматирования строк со
следующим функционалом:
– удаление всех слов из строки, длина которых меньше n букв;
– замена всех цифр в строке на знак «*»;
– вставка по одному пробелу между всеми символами в строке;
– сортировка слов по размеру;
– сортировка слов в лексикографическом порядке.
Примечание. Разделители слов можно задавать отдельно. По
умолчанию в качестве разделителя принимается только символ
пробела.
'''
None
```

```python
import re
```

```python
class StringFormatter:
    __slots__ = ['__line', '__split_char']
    
    def __init__ (self, line, split_char = ' '):
        if not line.strip() == "":
            self.__line = line
        else:
            print("The line is empty!")
        if len(split_char) == 1:
            self.__split_char = split_char
        else:
            self.__split_char = ' '
            
    def set_line(self, new_line):
        if not new_line.strip() == "":
            self.__line = new_line
        else:
            print("The line is empty!")
    
    def remove_words_len(self, num):
        if num < 1:
            print("Invalid length specified!")
            return
        
        result_words = []
        str_split = self.__line.split(self.__split_char)
        
        for word in str_split:
            if len(word) >= num:
                result_words.append(word)
                
        self.__line = ' '.join(result_words) 
        return self.__line
    
    def replace_num(self):
        self.__line = re.sub('\d', '*', self.__line)
        return self.__line
    
    def set_space(self):
        count_of_iter = 0
        
        for i in range(len(self.__line)-1):
            if not self.__line[i+count_of_iter] == ' ' and not self.__line[i+1+count_of_iter] == ' ':
                new = list(self.__line)
                new.insert(i+1+ count_of_iter, " ")
                self.__line = ''.join(new)
                count_of_iter += 1
                
        return self.__line
    
    def sort_by_length(self):
        str_split = self.__line.split(self.__split_char)
         
        for i in range(len(str_split)-1):
            for j in range(len(str_split)-i-1):
                if len(str_split[j]) > len(str_split[j+1]):
                    str_split[j], str_split[j+1] = str_split[j+1], str_split[j]
                    
        self.__line = ' '.join(str_split)
        return self.__line
    
    def sort_by_lexico(self):
        new_line = sorted((self.__line.lower()).split(self.__split_char))
        result = []
        
        for word_new_line in new_line:
            for word_line in (self.__line).split(self.__split_char):
                if word_line.lower() == word_new_line:
                    result.append(word_line)
                    self.__line = re.sub(r"\b" + re.escape(word_line) + r"", '', self.__line, 1)
                    break

        self.__line = " ".join(result)
        return self.__line
```

```python
class StringFrame(wx.Frame):
    ''' Класс окна работы со строками '''

    def __init__(self):
        wx.Frame.__init__(
            self, None, -1, "Обработка строк", size=(400, 320))

        panel = wx.Panel(self, -1)
        wx.StaticText(panel, -1, "Строка:", pos=(5, 20))
        self.entry_text = wx.TextCtrl(
            panel, -1, "", size=(300, -1), pos=(70, 20))
        wx.StaticText(panel, -1, "Результат:", pos=(5, 230))
        self.result_text = wx.TextCtrl(
            panel, -1, "", size=(300, -1), pos=(70, 230))

        self.sc = wx.SpinCtrl(panel, -1, "", (295, 55), (40, -1))
        self.sc.SetRange(1, 20)
        self.sc.SetValue(5)

        self.remove_smaller = wx.CheckBox(panel, -1, "Удалить слова размером меньше",
                    (70, 60), (250, 20))
        wx.StaticText(panel, -1, "букв", pos=(340, 62))
        self.replase_num = wx.CheckBox(panel, -1, "Заменить все цифры на *", (70, 80), (220, 20))
        self.space_in_char = wx.CheckBox(panel, -1, "Вставить пробелы между символами",
                    (70, 100), (280, 20))
        self.sort_checkbox = wx.CheckBox(panel, -1, "Сортировать слова в строке",
                                         (70, 120), (220, 20))

        self.radio_by_size = wx.RadioButton(
            panel, -1, "По размеру", (100, 140), (150, 20))
        self.radio_by_lex = wx.RadioButton(
            panel, -1, "Лексикографически", (100, 160), (150, 20))
        self.radio_by_size.Disable()
        self.radio_by_lex.Disable()

        format_button = wx.Button(panel, -1, "Форматирование",
                                  size=(300, 30), pos=(70, 190))

        self.Bind(wx.EVT_BUTTON, self.on_format_click, format_button)
        self.Bind(wx.EVT_CHECKBOX, self.on_check, self.sort_checkbox)

    def on_format_click(self, event):
        input_str = self.entry_text.Value
        sf = StringFormatter(input_str)
        if self.remove_smaller.IsChecked():
            if int(self.sc.GetValue()) > 0:
                input_str = sf.remove_words_len(int(self.sc.GetValue()))
                sf.set_line(input_str)
            else:
                wx.MessageBox('Указано неверное число.', 'Info', wx.OK | wx.ICON_INFORMATION)
        
        if self.replase_num.IsChecked():
            input_str = sf.replace_num()
            sf.set_line(input_str)
        
        if self.sort_checkbox.IsChecked():
            if self.radio_by_size.GetValue():
                input_str = sf.sort_by_length()
                sf.set_line(input_str)
            elif self.radio_by_lex.GetValue():
                input_str = sf.sort_by_lexico()
                sf.set_line(input_str)
            else:
                wx.MessageBox('Не указан метод сортировки.', 'Info', wx.OK | wx.ICON_INFORMATION)                
        
        if self.space_in_char.IsChecked():
            input_str = sf.set_space()
            sf.set_line(input_str)

        self.result_text.Value = input_str

    def on_check(self, event):
        if self.sort_checkbox.IsChecked():
            self.radio_by_size.Enable()
            self.radio_by_lex.Enable()
        else:
            self.radio_by_size.Disable()
            self.radio_by_lex.Disable()
```
